```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/process_v1/process_data_3.csv')
```

```python
df
```

```python
df = df[df['street'].isnull() == False]
```

```python
def concat(a, b):
    return f'{a} - {b}'
df['ward'] = df.apply(lambda x: concat(x['district'], x['ward']), axis = 1)
df['street'] = df.apply(lambda x: concat(x['ward'], x['street']), axis = 1)
```

```python
df
```

```python
df = df.reset_index(drop = True)
df.to_csv('../data/process_v1/process_data_4.csv', index = False)
```
