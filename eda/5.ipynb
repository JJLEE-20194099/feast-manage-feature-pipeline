```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/process_v1/process_data_2.csv')
```

```python
df.head(2)
```

```python
df['endWidth'] = df['endWidth'].replace(0, np.nan)
df['frontRoadWidth'] = df['frontRoadWidth'].replace(0, np.nan)
df['numberOfFloors'] = df['numberOfFloors'].replace(0, np.nan)

df['numberOfBathRooms'] = df['numberOfBathRooms'].replace(0, np.nan)
df['numberOfBedRooms'] = df['numberOfBedRooms'].replace(0, np.nan)
df['numberOfKitchens'] = df['numberOfKitchens'].replace(0, np.nan)
df['numberOfGarages'] = df['numberOfGarages'].replace(0, np.nan)
df['certificateOfLandUseRight'] == (df['certificateOfLandUseRight'] == 'yes').astype(np.int32)

```

```python
df['numberOfFloors'].value_counts()
```
